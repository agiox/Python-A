### Παράμετροι και ορίσματα
2.1 Σας δίνεται μια χρονική διάρκεια που αποτελείται από ώρες, λεπτά και δευτερόλεπτα, σαν συμβολοσειρά με τη μορφή `ωω:λλ:δδ`, π.χ. 05:33:07. 

Γράψτε κώδικα που να μετατρέπει τη χρονική διάρκεια σε έναν δεκαδικό αριθμό, όπου το ακέραιο μέρος είναι οι ώρες και τα δεκαδικά του ψηφία υπολογίζονται από τα λεπτά και τα δευτερόλεπτα που δόθηκαν. 

Υπενθυμίζεται ότι μία ώρα αποτελείται από 60 πρώτα λεπτά ή 3600 δευτερόλεπτα, οπότε το `01:30:00` αντιστοιχεί στον δεκαδικό 1.5, ενώ το `02:45:00` στον 2.75.

Θεωρήστε ότι οι χρόνοι που δίνονται έχουν _πάντα_ δύο ψηφία για τις ώρες, τα λεπτά και τα δευτερόλεπτα.

Χρησιμοποιείστε τουλάχιστον πέντε δεκαδικά ψηφία για την απεικόνιση της ώρας σαν δεκαδικό αριθμό.

```python
# Λύση 2.1


def hms_to_hours(hms):
    hours = int(hms[:2])
    minutes = int(hms[3:5])
    seconds = int(hms[6:])
    
    return hours + minutes / 60 + seconds / 3600


times_hms = ["02:40:08", "01:58:45", "10:09:57", "12:01:22", "04:45:33"]
for time in times_hms:
    hours = hms_to_hours(time)
    print(f"Ο χρόνος {time} αντιστοιχεί σε {hours:.6f} ώρες.")
```

## Προαιρετικά ορίσματα, προκαθορισμένες τιμές συναρτήσεων
2.2 Γράψτε κώδικα που να ορίζει μια συνάρτηση `accurate_enough(num_a, num_b, percent=0.1)` η οποία δέχεται ως υποχρεωτικά ορίσματα δύο θετικούς αριθμούς, `num_a`, `num_b` και προαιρετικά μια παράμετρο `percent` με προκαθορισμένη τιμή `0.1`. Η συνάρτηση να υπολογίζει το πόσο κοντά είναι οι δύο αριθμοί σαν ποοστό της τιμής του δεύτερου αριθμού. Για να το υπολογίσουμε αυτό, χρειάζονται δύο βήματα.

Α. Υπολογίζουμε τη _θετική_ διαφορά των δύο αριθμών σαν ποσοστό του δεύτερου αριθμού. 

Β. Τη διαφορά που υπολογίσαμε στο πρώτο βήμα τη διαιρούμε με τον δεύτερο αριθμό. Για να το μετατρέψουμε σε ποσοστό, πολλάπλασιάζουμε το αποτέλεσμα της διαίρεσης με το 100.

Γ. Αν ο αριθμός που προκύπτει είναι μικρότερος ή ίσος με την προαιρετική παράμετρο (που, αν δεν δώσουμε άλλη τιμή, ελέγχει για ακρίβεια 0.1%), τότε η συνάρτηση να επιστρέφει την τιμή `True`, αλλιώς την τιμή `False`.

Καλέστε τη συνάρτηση για διάφορες τιμές ορισμάτων έτσι ώστε να επιδεικνύεται η χρήση του προαιρετικού ορίσματος.

```python
# Λύση 2.2
def accurate_enough(num_a, num_b, percent=0.1):
    if num_a > num_b:
        diff = num_a - num_b
    else:
        diff = num_b - num_a
    perc_diff = diff / num_b * 100
    
    if perc_diff > percent:
        return False
    else:
        return True

print("Υπολογισμός για τους αριθμούς 10 και 12 με την προκαθορισμένη ακρίβεια: ", accurate_enough(10, 12))
print("Υπολογισμός για τους αριθμούς 1000 και 999 με την προκαθορισμένη ακρίβεια: ", accurate_enough(1000, 999))
print("Υπολογισμός για τους αριθμούς 999 και 1000 με την προκαθορισμένη ακρίβεια: ", accurate_enough(999, 1000))
print("Υπολογισμός για τους αριθμούς 4000 και 3999 με ακρίβεια 0.05%: ", accurate_enough(4000, 3999, percent=0.05))
print("Υπολογισμός για τους αριθμούς 31790 και 32767 με ακρίβεια 3%: ", accurate_enough(31784, 32767, percent=3))
print("Υπολογισμός για τους αριθμούς 42.42 και 41.99 με ακρίβεια 1%: ", accurate_enough(42.42, 41.99, percent=1))
```

2.3 Γράψτε κώδικα που να ορίζει μια συνάρτηση `speed_stats(*args)` η οποία δέχεται οποιονδήποτε αριθμό από ορίσματα συμβολοσειρών της μορφής `ωω:λλ:δδ` (όπως στην άσκηση 2.1).
Θεωρήστε ότι οι χρόνοι αφορούν διαδρομές μαραθωνίου, δηλαδή απόσταση 42.195 χιλιομέτρων.
Η συνάρτηση θέλουμε να επιστρέφει τα ακόλουθα: το πλήθος των ορισμάτων, τη μεγαλύτερη μέση ταχύτητα της διαδρομής, την ελάχιστη ταχύτητα, καθώς και τη μέση τιμή της ταχύτητας για όλες τις διαδρομές. 

Για τη μετατροπή των ορισμάτων σε δεκαδικούς, χρησιμοποιήστε τη συνάρτηση που γράψατε για την άσκηση 2.1.

Αν δεν δοθεί κανένα όρισμα, η συνάρτηση να επιστρέφει την τιμή `0` σε κάθε ένα από αυτά που ζητάμε να μας επιστρέφει. 
Καλέστε τη συνάρτηση με διαφορετικά πλήθη χρονικών διαστημάτων, ώστε να φανεί η λειτουργία της. Δοκιμάστε να την καλέσετε και χωρίς ορίσματα, ώστε να επιβεβαιώσετε ότι λειτουργεί σωστά.


```python
# Λύση

def hms_to_hours(hms):
    hours = int(hms[:2])
    minutes = int(hms[3:5])
    seconds = int(hms[6:])
    
    return hours + minutes / 60 + seconds / 3600

def speed_stats(*args):
    distance = 42.195

    max_speed, min_speed, mean_speed = 0, 1000, 0

    run_count = len(args)
    for race in args:
        speed = distance / hms_to_hours(race)
        mean_speed += speed
        if speed > max_speed:
            max_speed = speed
        if speed < min_speed:
            min_speed = speed
    
    if run_count > 0:
        mean_speed = mean_speed / run_count
    else:
        mean_speed = 0
        min_speed = 0

    return run_count, max_speed, min_speed, mean_speed


print(speed_stats("03:00:00", "04:00:00", "05:00:00"))
print(speed_stats("02:34:56", "04:32:10", "05:43:21", "03:14:15", "02:22:22", "03:14:14", "03:45:09"))
print(speed_stats())  # χωρίς ορίσματα
```

Η συνάρτηση αφαιρεί από την hours τις 8 ώρες του ωραρίου και το αποτέλεσμα, θετικό ή αρνητικό, το προσθέτει στην overtime.  

Χρησιμοποιήστε assert για να διασφαλίσετε ότι η `hours` δεν έχει παράλογη τιμή, π.χ. δεν είναι μεγαλύτερη από 24 ή ότι δεν είναι μεγαλύτερη από το μέγιστο επιτρεπτό ωράριο (π.χ. 13 ώρες) (επιλογή σας).

Δημιουργήστε μια συνάρτηση `init_monthly_overtime()` η οποία θα καλείται στην αρχή κάθε μήνα. Αν δει ότι η overtime είναι αρνητική, ελέγχει αν είναι περισσότερες ή ίσες με το ωράριο μιας ημέρας (8 ώρες) και αν αν ναι, εμφανίζει ενημερωτικό μήνυμα στην οθόνη, χωρίς να πειράξει την τιμή. Αν είναι θετική, επιστρέφει την τιμή και μηδενίζει την overtime για τον επόμενη μήνα. Επίσης, αν είναι περισσότερες από 20 οι υπερωρίες να εμφανίζει ενημερωτικό μήνυμα.

Δημιουργήστε μια συνάρτηση `show_overtime()` που εμφανίζει τις τρέχουσες υπερωρίες.

Δοκιμάστε τις συναρτήσεις καλώντας τις με διαφορετικές τιμές ώστε να φανεί η λειτουργία τους.

```python
# Λύση 2.4
overtime = 0

def day_workhours(hours):
    global overtime

    assert hours <= 24, "Οι ώρες είναι περισσότερες από 24."
    diff_hours = hours - 8
    overtime += diff_hours

def init_monthly_overtime():
    global overtime
    if overtime < 0:
        if overtime < -8:
            print(f"Υπάρχει έλλειμμα άνω των 8 ωρών: {overtime}")
        return 
    else:
        month_overtime = overtime
        overtime = 0
        if month_overtime > 20:
            print(f"Προσοχή - πολλές υπερωρίες: {month_overtime}")
        return month_overtime

def show_overtime():
    print(f"Μηνιαίες υπερωρίες μέχρι στιγμής: {overtime}.")

day_workhours(10)
day_workhours(7)
day_workhours(8.5)
day_workhours(23)
day_workhours(23)
show_overtime()       
print(f"Σύνολο υπερωριών μήνα: {init_monthly_overtime()}")
day_workhours(1)
day_workhours(1)
show_overtime()       
print(f"Σύνολο υπερωριών μήνα: {init_monthly_overtime()}")
# day_workhours(30)  # Αυτό θα ενεργοποιήσει την assert
```

Η συνάρτηση ψάχνει να βρει αν η συμβολοσειρά string_a περιέχεται στην αρχή της συμβολοσειράς string_b. 

H pos είναι ακέραιος, προαιρετικό όρισμα με αρχική τιμή 0 και δείχνει τη θέση στην αρχική συμβολοσειρά string_b όπου αρχίζουμε το ψάξιμο για το string_a.

Υλοποιήστε τη συνάρτηση χρησιμοποιώντας αναδρομή.

Κάθε φορά η search_string καλείται με μικρότερο κομμάτι της string_b κατά ένα χαρακτήρα. Αρχίζετε ελέγχοντας αν ο πρώτος χαρακτήρας της πρώτης συμβολοσειράς είναι ίδιος με τον πρώτο χαρακτήρα της δεύτερης. Αν ναι, ελέγχετε αν ταιριάζουν και οι υπόλοποι αντίστοιχοι χαρακτήρες. Αν βρέθηκε το string_a, τυπώνετε τη θέση της δεύτερης συμβολοσειράς όπου ξεκινάει η πρώτη συμβολοσειρά.

Αν, για παράδειγμα, η πρώτη συμβολοσειρά είναι η ab και η δεύτερη είναι η zxcabwer, τότε τυπώνεται η θέση 3, όπου ξεκινάει το ab.

Αν δεν είναι ίδιοι οι πρώτοι χαρακτήρες των srting_a και string_b ή δε βρεθεί ολόκληρη η string_a στην αρχή της string_b συνεχίζουμε με τη συνάρτηση να καλεί τον εαυτό της κρατώντας ίδιο το string_a, αλλά από το string_b αφαιρούμε τον πρώτο χαρακτήρα αριστερά και με την pos αυξημένη κατά 1.

Η συνάρτηση σταματάει όταν η string_b φτάσει να έχει μικρότερο μήκος από τη string_a.

Καλέστε τη συνάρτηση για διάφορες τιμές ορισμάτων.

```python
# Λύση 2.5
def search_string(string_a, string_b, pos=0):
    if pos == 0:
        print(f"Searching for {string_a}.")

    if len(string_a) > len(string_b):
        return
    else:
        if string_a[0] == string_b[0]:
            match = True
            for i in range(1, len(string_a)):
                if string_a[i] == string_b[i]:
                    pass
                else:
                    match = False
                    break
            if i == len(string_a) - 1 and match == True:
                print(f"Found in position {pos}.")
        search_string(string_a, string_b[1:], pos + 1)
        return

search_string("ab", "zxcabwer")
search_string("abc", "zxczzzzabcwerz")
search_string("zz", "zxczzzzabcwerz")
```
